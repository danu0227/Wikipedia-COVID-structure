```python
def travel_getcount(adj, seeds, lang):
    target_adj = {}
    visited = set()
    vertices = set(seeds)
    hop = 0
    
    f = open('./kcc2021/3keywordcount_%s_qid.json'%lang, 'r')
    countmap = json.load(f)
    f.close()
    
    counts = []
    while True:
        print(len(vertices))
        counts.append([(v, countmap.get(v)) for v in vertices])
        hop+=1
        new_childs = set()
        for v in vertices:
            if v in visited or v == '':
                continue
            visited.add(v)
            child = adj.get(v)
            if child:
                target_adj[v] = child
                new_childs.update(child)
            else:
                target_adj[v] = []
        vertices.update(new_childs)
        if new_childs.issubset(visited):
            true_vertices = set(target_adj.keys())
            for k, v in target_adj.items():
                if not set(v).issubset(true_vertices):
                    target_adj[k] = list(set(v).intersection(true_vertices))
            return counts
```

```python
lang_count = {}
for lang in langs:
    print(lang)
    json_file = open('kcc2021/%s_m_qid.json' % lang, 'r')
    adjacency_dict = json.load(json_file)
    lang_count[lang] = travel_getcount(adjacency_dict, ['Q81068910', 'Q82069695', 'Q84263196'], lang)
```
